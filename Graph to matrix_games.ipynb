```python
import pandas as pd
import networkx as nx
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib

# Load the CSV file generated by C program
df = pd.read_csv('hex_game_results_13.csv')
df.shape
```

```python
# Limit to 10,000 rows(just for more speed processing)
df = df.iloc[:10000]

# Convert 'Moves' column to lists of integers
df['Moves'] = df['Moves'].apply(lambda x: list(map(int, x.split())))

# Check the processed moves
print("Processed Moves:")
print(df['Moves'].head())

# Assuming board size of 13x13 for the Hex game
BOARD_SIZE = 13

# Function to convert moves into a final board state
def get_final_board_state(moves, board_size):
    board = np.zeros((board_size, board_size), dtype=int)
    for i, move in enumerate(moves):
        player = 1 if i % 2 == 0 else 2
        row, col = move // board_size, move % board_size
        board[row, col] = player
    return board

# Add the final board state to the DataFrame
df['FinalBoard'] = df['Moves'].apply(lambda moves: get_final_board_state(moves, BOARD_SIZE))

# Check the final board states
print("Final Board State for the first game:")
print(df['FinalBoard'].iloc[0])

# Split the data into training and testing sets
# We have a 'Winner' column in the dataset for labels
X_train, X_test, Y_train, Y_test = train_test_split(df['FinalBoard'], df['Winner'], test_size=0.2, random_state=42)

# Check the split data
print(f"Training set size: {len(X_train)}")
print(f"Testing set size: {len(X_test)}")
```

```python
def encode_hex_board(hex_board, board_size):
    symbol_names = []
    for i in range(board_size):
        symbol_names.append("C:%d" % (i))  # Column symbols
        symbol_names.append("R:%d" % (i))  # Row symbols

    # Add player and wall symbols
    symbol_names.append("Player1")
    symbol_names.append("Player2")
    symbol_names.append("TouchingRedWall")
    symbol_names.append("TouchingBlueWall")
    

    # Debug: print the symbols
    print(f"Symbol names: {symbol_names}")
    
    number_of_nodes = board_size * board_size
    
    # Initialize the Graphs class with the updated symbol names
    graphs = Graphs(1, symbol_names=symbol_names, hypervector_size=128, hypervector_bits=2)

    # After initializing the Graphs object, check the symbol registration
    print("Registered symbols in the Graphs object:", graphs.symbol_id)

    # Initialize the graph structure
    graphs.set_number_of_graph_nodes(0, number_of_nodes)
    graphs.prepare_node_configuration()
    # Print the registered graph nodes to ensure they exist
    print(f"Registered graph nodes: {graphs.graph_node_id[0]}")

    #manually adding
    graphs.graph_node_id[0] = {}  # Initialize the dictionary for graph 0
    for node_id in range(number_of_nodes):
        graphs.graph_node_id[0][node_id] = node_id  # Manually register each node
    # Check node registration
    print(f"Registered graph nodes: {graphs.graph_node_id[0]}") 

    # Debug: print the board being processed
    print(f"Processing board:\n{hex_board}")
    
    # Fill in the graph nodes with features
    for row in range(board_size):
        for col in range(board_size):
            node_id = row * board_size + col
            cell_value = hex_board[row, col]
            
            # Debug: print each node value being processed
            print(f"Processing node: row={row}, col={col}, value={cell_value}")

            # Add node features (position and ownership)
            graphs.add_graph_node_feature(0, node_id, "C:%d" % col)  # Column symbol
            graphs.add_graph_node_feature(0, node_id, "R:%d" % row)  # Row symbol
            
            if cell_value == 1:
                graphs.add_graph_node_feature(0, node_id, "Player1")  # Player 1
            elif cell_value == 2:
                graphs.add_graph_node_feature(0, node_id, "Player2")  # Player 2
            
            # Add adjacency to wall features (optional)
            if row == 0 or row == board_size - 1:
                graphs.add_graph_node_feature(0, node_id, "TouchingRedWall")
            if col == 0 or col == board_size - 1:
                graphs.add_graph_node_feature(0, node_id, "TouchingBlueWall")
    
    graphs.prepare_edge_configuration()

    # Add edges between adjacent cells in the Hex grid
    for row in range(board_size):
        for col in range(board_size):
            node_id = row * board_size + col
            adjacent_cells = get_adjacent_cells(row, col, board_size)
            for adj_row, adj_col in adjacent_cells:
                adj_node_id = adj_row * board_size + adj_col
                # Add the edge using add_graph_node_edge
                graphs.add_graph_node_edge(0, node_id, adj_node_id, edge_type=1)
    
    graphs.encode()  # Encode the graph after adding all nodes and edges
    
    return graphs
def get_adjacent_cells(row, col, board_size):
    # Define adjacent cells in a hexagonal grid
    adjacent = []
    if row > 0:
        adjacent.append((row - 1, col))  # Up
        if col < board_size - 1:
            adjacent.append((row - 1, col + 1))  # Up-right
    if row < board_size - 1:
        adjacent.append((row + 1, col))  # Down
        if col > 0:
            adjacent.append((row + 1, col - 1))  # Down-left
    if col > 0:
        adjacent.append((row, col - 1))  # Left
    if col < board_size - 1:
        adjacent.append((row, col + 1))  # Right
    return adjacent
```

```python
from GraphTsetlinMachine.graphs import Graphs
# Encode the final board states into graphs for training and testing
def encode_all_games(board_states, board_size):
    encoded_graphs = []
    
    for board in board_states:
        # Encode the board into a graph
        encoded_graph = encode_hex_board(board, board_size)
        encoded_graphs.append(encoded_graph)
    
    return encoded_graphs
```

```python
# Encode training and testing data
encoded_train_graphs = encode_all_games(X_train_list, BOARD_SIZE)
encoded_test_graphs = encode_all_games(X_test_list, BOARD_SIZE)

# Check one of the encoded graphs
print("Encoded graph for the first training game:")
print(encoded_train_graphs[0])
```

```python
#training step
from GraphTsetlinMachine.tm import MultiClassGraphTsetlinMachine

# Initialize the Graph Tsetlin Machine
tm = MultiClassGraphTsetlinMachine(
    number_of_clauses=5000,  # Adjust based on your problem complexity
    T=10000,                 # Threshold
    s=3.9,                   # Specificity parameter
    depth=1,                 # Depth of decision-making
    message_size=128,        # Size of hypervector used in GTM
    message_bits=2,          # Bits per hypervector
    max_included_literals=32  # Maximum literals in each clause
)

# Training loop
epochs = 100  # Adjust the number of epochs as needed
for epoch in range(epochs):
    # Train the model on the encoded training graphs
    tm.fit(encoded_train_graphs, Y_train, epochs=1, incremental=True)
    
    # Calculate the accuracy on the training and testing data
    train_accuracy = 100 * (tm.predict(encoded_train_graphs) == Y_train).mean()
    test_accuracy = 100 * (tm.predict(encoded_test_graphs) == Y_test).mean()
    
    print(f"Epoch {epoch + 1}: Train accuracy = {train_accuracy:.2f}%, Test accuracy = {test_accuracy:.2f}%")
```
